```python
df = df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
})
df.head()
```

```python
df['senioridade'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Não Remoto',
    50: 'Híbrido',
    100: 'Totalmente Remoto'
})
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})
df['tamanho_empresa'].value_counts()
```

```python
!pip install iso3166
import iso3166

# Filtrar o DataFrame para o cargo de 'Data Scientist'
df_cientista_dados = df_limpo[df_limpo['cargo'] == 'Data Scientist']

# Calcular o salário médio por país para cientistas de dados
df_salario_medio_pais_ds = df_cientista_dados.groupby('empresa')['usd'].mean().reset_index()

# Função para converter códigos ISO-ALPHA-2 para ISO-3
def convert_iso2_to_iso3(iso2_code):
    try:
        return iso3166.countries.get(iso2_code).alpha3
    except KeyError:
        return None

# Aplicar a conversão à coluna 'empresa'
df_salario_medio_pais_ds['empresa_iso3'] = df_salario_medio_pais_ds['empresa'].apply(convert_iso2_to_iso3)

# Remover linhas onde a conversão falhou (países não encontrados)
df_salario_medio_pais_ds.dropna(subset=['empresa_iso3'], inplace=True)

# Criar o mapa mundi interativo
fig = px.choropleth(
    df_salario_medio_pais_ds,
    locations='empresa_iso3',  # Coluna com os códigos dos países (ISO-3)
    locationmode='ISO-3', # Define o modo de localização para códigos de 3 letras
    color='usd', # Coluna que define a cor do país (salário médio)
    hover_name='empresa', # Mostra o nome do país ao passar o mouse
    color_continuous_scale=px.colors.sequential.Plasma, # Escolha de escala de cores
    title='Salário Médio para Cientista de Dados por País (USD)'
)

fig.show()
```

```python
!pip install pycountry
import pycountry
import plotly.express as px

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
